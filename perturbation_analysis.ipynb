```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
```

```python
def load_csv(filename, skip_rows):
    df = pd.read_csv(filename, skiprows=skip_rows, header=0)
    df.drop(df.columns[0], axis=1, inplace=True)
    display(df)
    return df

def plot_market_share_trend():
    fig, ax = plt.subplots(figsize=(15, 8))
    
    filename = f'./experiment_data/perturbation.csv'
    df = load_csv(filename, 28)
    df_0 = df.iloc[:, 1::3]
    df_0.iloc[0:1000, 2] = None
    display(df_0.iloc[995:1005])
    
    ax.axhline(y=40, linestyle='--', color='red', label='even market share for 2 retailers')
    ax.axhline(y=27, linestyle='--', color='purple', label='even market share for 3 retailers')
    ax.plot(df_0['[ market-share ] of retailer 0'], label='Retailer 1')
    ax.plot(df_0['[ market-share ] of retailer 1'], label='Retailer 2')
    ax.plot(df_0['[ market-share ] of retailer third-retailer-id'], label='Retailer 3')
    
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    ax.set_ylabel('Market Shares')
    ax.set_xlabel('Time ticks')
    plt.title('Market shares of retailers')
    plt.plot()
```

```python
plot_market_share_trend()
```
