```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_csv(filename, skip_rows):
    df = pd.read_csv(filename, skiprows=skip_rows, header=0)
    df.drop(df.columns[0], axis=1, inplace=True)
    return df
```

```python
df = load_csv("/Users/austincheang/Desktop/Msc/ABM/ABM_Netlogo/experiment_data/E1_E3_2.csv", 30)
```

```python
df.columns
```

```python
retailer_0_price = df.filter(regex='^\[ price \] of retailer 0')
```

```python
retailer_1_price = df.filter(regex='^\[ price \] of retailer 1')
```

```python
retailer_0_price
```

```python
retailer_1_price
```

```python
def plot_price_evolve():
    sns.set_style("whitegrid", {'axes.grid': False})
    plt.figure(figsize=(8, 6))
    sns.lineplot(data=retailer_0_price["[ price ] of retailer 0.1"], label='Retailer_0');
    sns.lineplot(data=retailer_1_price["[ price ] of retailer 1.1"], label='Retailer_1');
    plt.axhline(y=30, color='r', linestyle='--', label='Unit Cost')
    plt.ylabel("Price")
    plt.xlabel("Time ticks")
    plt.title("Retailer Price evolution")

    # plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    # plt.savefig("./table_data/Retailer_price_evolution.jpg")
    plt.legend()
```

```python
plot_price_evolve()
```

```python
df = load_csv("/Users/austincheang/Desktop/Msc/ABM/ABM_Netlogo/experiment_data/E3_E2.csv", 30)
```

```python
retailer_0_profit = df.filter(regex='^\[ cumulative-profit \] of retailer 0')
retailer_1_profit = df.filter(regex='^\[ cumulative-profit \] of retailer 1')
```

```python
def plot_profit_evolve():
    plt.figure(figsize=(8, 6))
    sns.set_style("whitegrid", {'axes.grid': False})
    sns.lineplot(data=retailer_0_profit["[ cumulative-profit ] of retailer 0"], label='Retailer_0');
    sns.lineplot(data=retailer_1_profit["[ cumulative-profit ] of retailer 1"], label='Retailer_1');
    plt.ylabel("Cumulative Profit")
    plt.xlabel("Time ticks")
    plt.title("Retailer Profit")
    # plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
    plt.legend()
```

```python
plot_profit_evolve()
```
