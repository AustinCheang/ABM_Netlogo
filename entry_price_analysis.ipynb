```python
import numpy as np
import pandas as pd
```

```python
def load_csv(filename, skip_rows):
    df = pd.read_csv(filename, skiprows=skip_rows, header=0)
    df.drop(df.columns[0], axis=1, inplace=True)
    return df

def calculate_avg(df):
    df_even = df.iloc[:, ::2]
    display(df_even)
    even_row_avg = df_even.iloc[0].mean()
    print('AVG Cumulative Profit of retailer 0:', even_row_avg)
    
    df_odd = df.iloc[:, 1::2]
    display(df_odd)
    odd_row_avg = df_odd.iloc[0].mean()
    print('AVG Cumulative Profit of retailer 1:', odd_row_avg)
```

```python
df_all = load_csv("cumulative_profit.csv", 29)
df_all
```

```python
df_finals = load_csv("cumulative_profit_finals_only.csv", 24)
df_finals
```

```python
calculate_avg(df_finals)
```
