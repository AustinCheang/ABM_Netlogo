```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Tabel 3

```python
table_3 = pd.read_csv("./table_data/table_3/table_3.csv", header=None)
```

```python
def separate_retailer(table):
    strategies = ['S1', 'S2', 'S3', 'S4', 'S5', 'S6']
    retailer_1 = table.loc[:, [x for x in list(range(12)) if x % 2 == 0]]
    retailer_2 = table.loc[:, [x for x in list(range(12)) if x % 2 != 0]]

    retailer_1.columns = strategies
    retailer_1.index = strategies
    retailer_2.columns = strategies
    retailer_2.index = strategies
    
    return retailer_1, retailer_2
    
```

```python
retailer_1, retailer_2 = separate_retailer(table_3)
```

```python
retailer_2
```

```python
def plot_heatmap_for_retailer_x(title, retailer_df, path):
     plt.figure(figsize=(12, 6))
     sns.heatmap(retailer_df, cmap='BuPu', annot=True, fmt='.2f');
     plt.xlabel(f'Retailer 2 Strategies', fontsize=12)
     plt.ylabel(f'Retailer 1 Strategies', fontsize=12)
     plt.title(title, fontsize=16)
     plt.savefig(f'{path}/{title}')
```

```python
plot_heatmap_for_retailer_x('Retailer 1 Payoff', retailer_1, path='./table_data/table_3');
plot_heatmap_for_retailer_x('Retailer 2 Payoff', retailer_2, path='./table_data/table_3');
```

```python
# Average of two payoffs
retailer_avg = (retailer_1 + retailer_2) / 2
```

```python
plot_heatmap_for_retailer_x('Average Payoff of the market', retailer_avg)
```

# Table 6

```python
table_6 = pd.read_csv("./table_data/table_6/table_6.csv", header=None)
```

```python
table_6.shape
```

```python
def separate_entry_price(table):
    strategies = ['E1', 'E2', 'E3']
    retailer_1 = table.loc[:, [x for x in list(range(table.shape[1])) if x % 2 == 0]]
    retailer_2 = table.loc[:, [x for x in list(range(table.shape[1])) if x % 2 != 0]]

    retailer_1.columns = strategies
    retailer_1.index = strategies
    retailer_2.columns = strategies
    retailer_2.index = strategies
    
    return retailer_1, retailer_2
    
```

```python
retailer_1_entry, retailer_2_entry = separate_entry_price(table_6)
```

```python
retailer_2_entry
```

```python
plot_heatmap_for_retailer_x('Retailer 1 Payoff', retailer_1_entry, path="./table_data/table_6");
plot_heatmap_for_retailer_x('Retailer 2 Payoff', retailer_2_entry, path="./table_data/table_6");
```

```python
retailer_sum = retailer_1_entry + retailer_2_entry
```

```python
plot_heatmap_for_retailer_x('Market Sum Payoff', retailer_sum, path="./table_data/table_6");
```
